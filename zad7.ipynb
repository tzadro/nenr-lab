```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Zadatak
Varijabla $s$ određuje "skalu", tj. koliko je udaljenost bitna. Ako imamo više ulaza svaki $s$ će kontrolirati skalu za pojedinu dimenziju.

```python
def y(x, w, s):
    return 1. / (1 + abs(x - w) / abs(s))

w = 2
x = np.linspace(-8, 10, 100, endpoint=True)

plt.plot(x, y(x, w, 1), label='s = 1', c='r')
plt.plot(x, y(x, w, 0.25), label='s = 0.25', c='g')
plt.plot(x, y(x, w, 4), label='s = 4', c='b')
plt.xlim(-8, 10)
plt.legend()
plt.show()
```

### 2. Zadatak
Postoji uzorak, ali razredi nisu linearno odvojivi.

```python
input_file = './materijali/zad7-datasets/zad7-dataset.txt'
num_examples = 64

x = [[], [], []]
y = [[], [], []]

with open(input_file, 'r') as f:
    for _ in range(num_examples):
        line = np.fromstring(f.readline(), count=5, sep='\t')
        z_ = np.argmax(line[2:])
        x[z_].append(line[0])
        y[z_].append(line[1])

plt.scatter(x[0], y[0], label='A', marker='o', color='r')
plt.scatter(x[1], y[1], label='B', marker='s', color='g')
plt.scatter(x[2], y[2], label='C', marker='^', color='b')
plt.legend()
plt.show()
```
