```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Zadatak
Varijabla $s$ određuje "skalu", tj. koliko je udaljenost bitna. Ako imamo više ulaza svaki $s$ će kontrolirati skalu za pojedinu dimenziju.

```python
def y(x, w, s):
    return 1. / (1 + abs(x - w) / abs(s))

w = 2
x = np.linspace(-8, 10, 100, endpoint=True)

plt.plot(x, y(x, w, 1), label='s = 1', c='r')
plt.plot(x, y(x, w, 0.25), label='s = 0.25', c='g')
plt.plot(x, y(x, w, 4), label='s = 4', c='b')
plt.xlim(-8, 10)
plt.legend()
plt.show()
```

### 2. Zadatak
Postoji uzorak, ali razredi nisu linearno odvojivi.

```python
def plot_dataset(limit=True):
    input_file = './materijali/zad7-datasets/zad7-dataset.txt'
    num_examples = 64

    x = [[], [], []]
    y = [[], [], []]

    with open(input_file, 'r') as f:
        for _ in range(num_examples):
            line = np.fromstring(f.readline(), count=5, sep='\t')
            z_ = np.argmax(line[2:])
            x[z_].append(line[0])
            y[z_].append(line[1])

    plt.figure(figsize=(8, 8))
    if limit:
        plt.ylim(0, 1)
        plt.xlim(0, 1)
    plt.scatter(x[0], y[0], label='A', marker='o', color='r')
    plt.scatter(x[1], y[1], label='B', marker='s', color='g')
    plt.scatter(x[2], y[2], label='C', marker='^', color='b')
```

```python
plot_dataset()
plt.legend()
plt.show()
```

### 3. Zadatak
Kako svaki neuron prvog sloja definira jednu točku u prostoru od koje mjeri udaljenost ulaznog primjera, u ovom slučaju uzeo bih 8 neurona u prvom sloju i $w$ od svakoga bi stavio u središte jednog od krugova. Težine iz svakog neurone koje idu u sljedeći sloj imale bi veliku vrijednost samo na onim težinama koje vode na izlazni neuron odgovarajuće klase.

### 4. Zadatak

```python
params = [0.88, 0.60, -0.09, -7.47, 0.15, 0.24, 0.04, -4.74, -1.77, -0.67, -0.00, 1.05, 0.06, 0.23, 0.18, -0.08, -1.81, -2.90, 0.53, 0.01, 0.91, 0.58, -0.10, -5.20, 0.08, 0.31, -0.02, 3.45, 0.13, 0.36, -0.04, 2.33]
layers = [2, 8, 3]

w1 = []
w2 = []
s = []
for i in range(layers[1]):
    start = i * layers[0] * 2
    w1.append(params[start])
    w2.append(params[start + 1])
    s.append('(' + str(params[start + 2]) + ', ' + str(params[start + 3]) + ')')

def plot_first_layer(w1, w2, s, limit):
    plot_dataset(limit)
    plt.scatter(w1, w2, label='w', marker='*', color='m')

    ax = plt.gca()
    for i, txt in enumerate(s):
        ax.annotate(txt, (w1[i],w2[i]))

    plt.legend()
    plt.show()
    
plot_first_layer(w1, w2, s, False)
plot_first_layer(w1, w2, s, True)
```
